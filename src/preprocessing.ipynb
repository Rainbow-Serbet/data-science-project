```python
import pandas as pd
import numpy as np
import matplotlib as plt
import time
import os
import warnings
warnings.filterwarnings(action='ignore') #mean 경고창 안뜨게
```

```python
#1년

#기호 식품과 생필품 저장 리스트
favor_product = ["빵", "즉석식품", "커피", "스낵과자", "비스킷", "아이스크림", "파운데이션", "드레싱",
                 "전복", "커피크림", "잼", "탄산음료", "클렌징크림"]

ess_product = ["치약", "화장지", "칫솔", "콩", "샴푸", "냉동식품", "라면", "분유", "고등어", "햄", "로션", "면도기", "된장",
                "참기름", "생수", "식용유", "케찹", "설탕", "밀가루", "고추장", "소금", "쌀"]

favor_list = []
ess_list = []

for month in range(1, 13, 1):
    for day in range(1, 32, 1):
        try:
            data = pd.read_csv(f'Pr_Data/2017-{month:02d}-{day:02d}.csv', sep = ',', engine = 'python', encoding = 'cp949')
        except FileNotFoundError as e:
            pass
        
        # df = data.drop(["good_id", "pum_id", "good_name", "discount_price", "benifit"], axis=1)
        # df2 = df.dropna(axis=0)
        
        data_tbl = pd.DataFrame(data, columns = ('collect_day', 'pum_name', 'sales_price'))
        data_tbl = data_tbl.astype({'pum_name':'string'})
        
        fav_pro = data_tbl[data_tbl['pum_name'].str.contains('|'.join(favor_product))]
        ess_pro = data_tbl[data_tbl['pum_name'].str.contains('|'.join(ess_product))]
        
        fav_pro2 = fav_pro.mean()
        ess_pro2 = ess_pro.mean()
        
        #만약 날짜와 평균값이 NaN일 때 - 파일이 비어있을 때
        data_sum = fav_pro2.isnull().sum()
        
        if data_sum > 1:
            fav_pro2 = fav_pro2.fillna(0)
            ess_pro2 = ess_pro2.fillna(0)
            
            favor_list.append(fav_pro2)
            ess_list.append(ess_pro2)
        else:
            fav_pro2 = fav_pro2.astype(float)
            ess_pro2 = ess_pro2.astype(float)
        
            fav_pro2 = fav_pro2.round()
            ess_pro2 = ess_pro2.round()
            
            favor_list.append(fav_pro2)
            ess_list.append(ess_pro2)
            
        fav_pro2['collect_day'] = f"2018-{month}-{day}"
        ess_pro2['collect_day'] = f"2018-{month}-{day}"
    
fav_tbl = pd.DataFrame(favor_list, columns = ['collect_day', 'sales_price'])
ess_tbl = pd.DataFrame(ess_list, columns = ['collect_day', 'sales_price'])

fav_tbl.columns = ['날짜', '기호식품']
ess_tbl.columns = ['날짜', '생필품']

merge_tbl = pd.merge(fav_tbl, ess_tbl, on='날짜')

#행 열 전환
# merge_tbl = merge_tbl.transpose()
# merge_tbl.rename(columns=merge_tbl.iloc[0], inplace=True)
# merge_tbl = merge_tbl.drop(merge_tbl.index[0])

#데이터 기록하면 오른쪽으로 쭉 되는게 아니라 아래로 기록됨 - 문제
#인코딩 utf-8로 할 것

merge_df = pd.DataFrame(merge_tbl, columns = ['기호식품', '생필품'])

if os.path.exists('revise_tbl.csv'):
    merge_tbl.to_csv('revise_tbl.csv', encoding='utf-8', mode='a', header=False, index=False)
else:
    merge_tbl.to_csv('revise_tbl.csv', encoding='utf-8', mode='a', header=True, index=False)
    
```
