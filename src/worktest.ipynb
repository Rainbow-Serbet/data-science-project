```python
import datetime
import pandas as pd
import numpy as np
import matplotlib as plt
import time
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.filterwarnings("ignore", category=np.VisibleDeprecationWarning)

fav_avg_data = []
ess_avg_data = []
day_data = []

def cleaning(x) :
    if str(x).find("(") != -1 :
        return str(x).split("(")[1].replace(")","") 
    return str(x)

def d_change(x) :
    if str(x)[5]=="0":
        val=list(x)
        val[5]=""
        return "".join(val)
    return str(x)

def d_change2(x) :
    if str(x)[len(str(x))-2]=="0":
        val=list(x)
        val[len(val)-2]=""
        return "".join(val)
    return str(x)

def d_change3(x) :
    if str(x)[len(str(x))-5]=="0":
        val=list(x)
        val[len(val)-5]=""
        return "".join(val)
    return str(x)

#영서 2014 - 2016
yung_url = "./final_data/yung/revise_tbl.csv"
#문수 2017 ~ 2019
munsu_url = "./final_data/munsu/revise_tbl_2017_2019.csv"
#동협 2020 ~ 2022
dong_fav_url = "./final_data/fav_tbl2.csv"
dong_ess_url = "./final_data/ess_tbl2.csv"

#cos  - 코스피 저장 형태 오류남 Cod 맞음 .
cos = "./final_data/cos.csv"
#bond  - 채권
bond = "./final_data/bond.csv"


data_first = pd.read_csv(yung_url,sep=",",engine="python",encoding="utf-8",error_bad_lines=False, warn_bad_lines=False)
data_second = pd.read_csv(munsu_url,sep=",",engine="python",encoding="utf-8",error_bad_lines=False, warn_bad_lines=False)

data_third_1 = pd.read_csv(dong_fav_url,sep=",",engine="python",encoding="cp949",error_bad_lines=False, warn_bad_lines=False)
data_third_2 = pd.read_csv(dong_ess_url,sep=",",engine="python",encoding="cp949",error_bad_lines=False, warn_bad_lines=False)

data_third_1=data_third_1.drop([data_third_1.columns[0]],axis=1)
data_third_2=data_third_2.drop([data_third_2.columns[0],data_third_2.columns[1]],axis=1)

data_third_1.columns = ['날짜','기호식품']
data_third_2.columns = ['생필품']
# print(data_third_1.head())
# print(data_third_2.head())

#코스피 및 채권
cos_data = pd.read_csv(cos,sep=",",engine="python",encoding="utf-8",error_bad_lines=False, warn_bad_lines=False)
bond_data = pd.read_csv(bond,sep=",",engine="python",encoding="utf-8",error_bad_lines=False, warn_bad_lines=False)

# cos
# 날짜	종가	오픈	고가	저가	거래량	변동 %

# bond
# 날짜	종가	오픈	고가	저가	변동 %
cos_data=cos_data.drop(columns=['오픈','고가','저가','거래량','변동 %'],axis=1)
bond_data=bond_data.drop(columns=['오픈','고가','저가','변동 %'],axis=1)

# print(cos_data)
# print(bond_data)

# 역순
cos_data=cos_data.sort_index(ascending= False)
bond_data=bond_data.sort_index(ascending= False)

data_third=pd.concat([data_third_1,data_third_2],axis=1)
# print(data_third)

data=pd.concat([data_first,data_second,data_third],axis=0)

# 동협 파일 -> 날짜 없는부분 삭제
data=data.reset_index(drop=True)
data=data[data.index <= 3224]

data.to_csv('./final_data/final_data1.csv', encoding='utf-8', mode='w', header=False)

# 2014-1-1
# print(data)
# 2014.1.1
# print(cos_data)
# 2014년 01월 01일
# print(bond_data)
# print(bond_data)



cos_data['날짜']=cos_data['날짜'].apply(lambda x:x.replace(".","-"))
bond_data['날짜']=bond_data['날짜'].apply(lambda x:x.replace(" ",""))
# print(bond_data)
bond_data['날짜']=bond_data['날짜'].apply(lambda x:x.replace("년","-"))
bond_data['날짜']=bond_data['날짜'].apply(lambda x:x.replace("월","-"))
bond_data['날짜']=bond_data['날짜'].apply(lambda x:x.replace("일",""))
bond_data['날짜']=bond_data['날짜'].apply(d_change)
bond_data['날짜']=bond_data['날짜'].apply(d_change2)

data['날짜']=data['날짜'].apply(d_change3)
data['날짜']=data['날짜'].apply(d_change2)

data.to_csv('./final_data/final_data4.csv', encoding='utf-8', mode='w', header=False)



# bond 일 뒷자리 2자리 .
# print(bond_data)
# print(data)
# print(cos_data)

# data_test1=pd.merge(data,cos_data,on="날짜",how="inner")
# print(data_test1)
# data_test2=pd.merge(data,cos_data,on="날짜",how="outer")
# print(data_test2)


# data.to_csv('./final_data/merge_cos.csv', encoding='utf-8', mode='w', header=False)

# 정상
# print(data)
# data.set_index("날짜")
# print(data)

# print(bond_data)
# bond_data.set_index("날짜")
# print(bond_data)
# bond_data.to_csv('./final_data/bond_data.csv', encoding='utf-8', mode='w', header=False)

# data=pd.merge(data,bond_data,on="날짜",how="outer")

######################################################################
# data=pd.merge(data,bond_data,on="날짜",how="inner")
######################################################################

#cos  2172
#bond 2939
# print(cos_data)
# print(bond_data)

data_cos_bond=pd.merge(cos_data,bond_data,on="날짜",how="inner")
#2171
print(data_cos_bond)

data_cos_bond.to_csv('./final_data/final_data2.csv', encoding='utf-8', mode='w', header=False)

data=pd.merge(data,data_cos_bond,on="날짜",how="inner")
# print(data)

# 날짜     기호식품      생필품      종가_x   종가_y
data.columns = ["날짜","기호식품","생필품","코스피","채권"]

data.to_csv('./final_data/final_data.csv', encoding='utf-8', mode='w', header=True)

#전치행렬
data=data.T
print(data)

data.to_csv('./final_data/final_dataT.csv', encoding='utf-8', mode='w', header=True)



# 종가 1967.19
# print(data)

# 문수 영서 내파일 합치고 
# 결측치 처리 -> 전날 다음날 합쳐서  평균내서 채워주기 
# 코스피 , 채권 
#2014-10-01
#2014-10-11
```
